```python
%pylab inline
from sklearn.linear_model import SGDClassifier, LogisticRegression
```

```python
data = np.loadtxt('ex2data1.txt', delimiter=',')
X = data[:,:-1]
y = data[:,-1].reshape((-1,1))
X_pos = X[y[:,0] == 1]
X_neg = X[y[:,0] == 0]
```

```python
plt.plot(*X_pos.T,'k+',label='Admitted')
plt.plot(*X_neg.T,'ko',label='Not Admitted')
plt.xlabel('Exam 1 score')
plt.ylabel('Exam 2 score')
plt.legend(loc=1);
```

### Logistic Regression

```python
sgd = LogisticRegression(solver='newton-cg',
                         C=1e10,
                        max_iter=100)
sgd.fit(X,y.ravel())

x_fit = np.arange(X[:,0].min(), X[:,0].max())
y_fit = -(sgd.intercept_[0] + sgd.coef_[0,0]*x_fit)/sgd.coef_[0,1]

plt.plot(*X_pos.T,'k+',label='Admitted')
plt.plot(*X_neg.T,'ko',label='Not Admitted')
plt.plot(x_fit, y_fit, 'r-', label='Decision Boundary')
plt.xlabel('Exam 1 score')
plt.ylabel('Exam 2 score')
plt.legend(loc=1);
```

### Stochastic Gradient Descent Classification

```python
sgd = SGDClassifier(loss='log',
                    max_iter=150000,
                    penalty='none')
sgd.fit(X,y.ravel())

x_fit = np.arange(X[:,0].min(), X[:,0].max())
y_fit = -(sgd.intercept_[0] + sgd.coef_[0,0]*x_fit)/sgd.coef_[0,1]

plt.plot(*X_pos.T,'k+',label='Admitted')
plt.plot(*X_neg.T,'ko',label='Not Admitted')
plt.plot(x_fit, y_fit, 'r-', label='Decision Boundary')
plt.xlabel('Exam 1 score')
plt.ylabel('Exam 2 score')
plt.legend(loc=1);
```
