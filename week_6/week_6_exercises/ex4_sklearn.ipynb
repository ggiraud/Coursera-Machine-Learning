```python
%pylab inline
from scipy.io import loadmat
from sklearn.neural_network import MLPClassifier
```

```python
ex4data1 = loadmat('ex4data1.mat')
X,y = np.ascontiguousarray(ex4data1['X']), np.ascontiguousarray(ex4data1['y'])
# replace 10 class by 0
y[y[:,0] == 10, 0] = 0
```

```python
X_copy = X.copy()
np.random.shuffle(X_copy)
img = np.reshape(X_copy[:100], (10,10,20,20)).transpose((0,3,1,2)).reshape((200,200))
f = plt.imshow(img, cmap='gray');
```

### Multi Layer Perceptron

```python
clf = MLPClassifier(hidden_layer_sizes=(25),
                   activation='logistic',
                   solver='lbfgs',
                   alpha=1)
clf.fit(X,y.ravel())
clf.score(X,y.ravel())
```

```python
img = np.reshape(clf.coefs_[0].T.flatten(), (5,5,20,20)).transpose((0,3,1,2)).reshape((100,100))
plt.imshow(img, cmap='gray');
```
